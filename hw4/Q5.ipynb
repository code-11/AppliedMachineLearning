```python
import numpy as np
import pprint as pp
from PIL import Image

%matplotlib inline
def make_all_coords(pic_info):
    x_size=pic_info.shape[0]
    y_size=pic_info.shape[1]
    all_coords=np.transpose(np.mgrid[0:x_size,0:y_size])
    all_coords_fixed=np.reshape(all_coords,(-1,2))
    return all_coords_fixed
def setup():
    pil_im = Image.open('bay_of_naples.jpeg', 'r')
    pic_info=np.asarray(pil_im)
    all_coords_fixed=make_all_coords(pic_info)
    rnd_indexes=np.random.choice(all_coords_fixed.shape[0],5000,replace=False)
    rnd_coords=all_coords_fixed[rnd_indexes]
    rgbs=[]
    for coord in rnd_coords:
        rgb=pic_info[(coord[0],coord[1])]
        rgbs.append(rgb)
    rgbs=np.array(rgbs)
    return (rnd_coords,rgbs/255.0)
```

```python
from sklearn.ensemble import RandomForestRegressor
from matplotlib.pyplot import imshow 

coords,colors=setup()
model=RandomForestRegressor(random_state=0, n_estimators=100)
model.fit(coords,colors)
estimate=model.predict(make_all_coords())
estimate_pic=np.reshape(estimate,(1200,815,3))

```

```python
imshow(np.fliplr(np.rot90(estimate_pic,k=3)))
```

```python
from sklearn.ensemble import RandomForestRegressor
# from matplotlib.pyplot import imshow 
import matplotlib.pyplot as pyplot

# I=imread('./../images/pap.png');
# subplot(1,2,1);
# imshow(I);
# title('My Image'); 

def lots_of_depths():
    pil_im = Image.open('bay_of_naples.jpeg', 'r')
    pic_info=np.asarray(pil_im)
    depths=[1,2,3,5,10,15]
    for i in range(len(depths)):
        depth=depths[i]
        coords,colors=setup()
        model=RandomForestRegressor(random_state=0, n_estimators=1,max_depth=depth)
        model.fit(coords,colors)
        estimate=model.predict(make_all_coords(pic_info))
        estimate_pic=np.reshape(estimate,(1200,815,3))
        pyplot.figure(figsize = (25,20))
        pyplot.subplot(3,2,i)
        pyplot.imshow(np.fliplr(np.rot90(estimate_pic,k=3)))
        pyplot.title("Random forest with depth"+str(depth))
lots_of_depths()
```

```python
from sklearn.ensemble import RandomForestRegressor
import matplotlib.pyplot as pyplot

def lots_of_trees():
    pil_im = Image.open('bay_of_naples.jpeg', 'r')
    pic_info=np.asarray(pil_im)
    tree_nums=[1,3,5,10,100]
    for i in range(len(tree_nums)):
        num_tree=tree_nums[i]
        coords,colors=setup()
        model=RandomForestRegressor(random_state=0, n_estimators=num_tree,max_depth=7)
        model.fit(coords,colors)
        estimate=model.predict(make_all_coords(pic_info))
        estimate_pic=np.reshape(estimate,(1200,815,3))
        pyplot.figure(figsize = (25,20))
        pyplot.subplot(3,2,i)
        pyplot.imshow(np.fliplr(np.rot90(estimate_pic,k=3)))
        pyplot.title("Random forest with"+str(num_tree)+" tree(s)")
lots_of_trees()
```

```python
from sklearn.neighbors import KNeighborsRegressor

def knn():
    pil_im = Image.open('bay_of_naples.jpeg', 'r')
    pic_info=np.asarray(pil_im)
    coords,colors=setup()
    model=KNeighborsRegressor(n_neighbors=1)
    model.fit(coords,colors)
    estimate=model.predict(make_all_coords(pic_info))
    estimate_pic=np.reshape(estimate,(1200,815,3))
    pyplot.figure(figsize = (50,40))
    pyplot.imshow(np.fliplr(np.rot90(estimate_pic,k=3)))
    pyplot.title("KNN with k=1")
knn()
```

```python
from sklearn.ensemble import RandomForestRegressor
import sklearn.tree
from matplotlib.pyplot import imshow 
import os

coords,colors=setup()
model=RandomForestRegressor(random_state=0, n_estimators=100)
model.fit(coords,colors)
the_tree=model.estimators_[0]
sklearn.tree.export_graphviz(the_tree, max_depth=2, out_file='tree.dot') #print dotfile
os.system("dot -Tpng tree.dot -o tree.png" )
```
